```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', None)

sns.set_style("whitegrid")
```

```python
train_set = pd.read_json('train.json')
train_set.drop(columns=['id'], inplace=True)
train_set.sample(5)
```

```python
ing_col = train_set.ingredients.values
all_ings = np.concatenate(tuple(ing_col[i] for i in range(len(ing_col))))
ings=pd.DataFrame(all_ings, columns=["Ingredients"])
unique_in=ings.Ingredients.unique()
temp=pd.DataFrame(0, index= unique_in, columns= train_set.groupby(['cuisine']).count().index.values)

k=0
for i in range(len(train_set)):
    for j in range(len(train_set.ingredients[i])):
        temp[train_set["cuisine"][i]][train_set.ingredients[i][j]]=temp[train_set["cuisine"][i]][train_set.ingredients[i][j]]+1
temp.sample(5)
```

```python
temp['0 ingredient count'] = temp.isin([0]).sum(1)
temp.head(5)
```

```python
ms = [15, 16, 17, 18]
ns = [5, 10, 15, 20]
for m in ms:
    for n in ns:
        single_ingredients = temp[(temp['0 ingredient count'] > m) & (temp.iloc[:,:-1].sum(axis=1) > n)]
        print('number of ingredients present in no more than ', str(19-m),' cuisine types and with at least', str(n), ' recipe occurences: ', len(single_ingredients))
        print('has at least 1 ingredient for all cuisine types: ', str(np.all(single_ingredients.sum(axis=0).values > 0)))
```
