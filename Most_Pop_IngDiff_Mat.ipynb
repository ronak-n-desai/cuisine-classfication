```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
train_set = pd.read_json('train.json')
train_set.drop(columns=['id'], inplace=True)
train_set.sample(5)
```

```python
ing_col = train_set.ingredients.values
all_ings = np.concatenate(tuple(ing_col[i] for i in range(len(ing_col))))
unique_ings, counts = np.unique(all_ings, return_counts=True)
print('number of ingredients in dataset: ', len(all_ings))
print('unique number of ingredients in dataset: ', len(unique_ings))
```

```python
ings=pd.DataFrame(all_ings, columns=["Ingredients"])
unique_in=ings.Ingredients.unique()
```

```python
count_by_cuisine = train_set.groupby(['cuisine']).count()
count_by_cuisine
```

```python
temp=pd.DataFrame(0, index= unique_in, columns= count_by_cuisine.index.values)
temp
```

```python
k=0
for i in range(len(train_set)):
    for j in range(len(train_set.ingredients[i])):
        temp[train_set["cuisine"][i]][train_set.ingredients[i][j]]=temp[train_set["cuisine"][i]][train_set.ingredients[i][j]]+1
temp
```

```python
pop_ing = temp[temp.sum(axis=1)>1000]
display(pop_ing)
```

```python
Ab_diff_mat = pd.DataFrame(0,index = count_by_cuisine.index.values, columns = count_by_cuisine.index.values)

braz_minus_brit = np.abs(pop_ing.values[:,0] - pop_ing.values[:,1])
print("Absolute difference Brazilian and Brittish:", braz_minus_brit)
print("Sum of Absolute differences Brazilian and Brittish:", sum(braz_minus_brit))

display(Ab_diff_mat)
```

```python
for i in np.arange(len(count_by_cuisine)):
    for j in np.arange(len(count_by_cuisine)):
        Ab_diff_mat.values[i,j] = sum(np.abs(pop_ing.values[:,i] - pop_ing.values[:,j]))
        
display(Ab_diff_mat/Ab_diff_mat.values.max())
```
